```python
#Evolutionary Algorithm
"""The following evolutionary program is to find the optimum solution for bin-packing problem that involves
the evolution process"""
import random
import numpy as np
import matplotlib.pyplot as plt

def spawn_candidate_solutions(start_value=1, end_value=101, length_of_candidate=500, candidate_soln_population_size = 100):
    """Method Name:spawn_candidate_solutions()
    Functionality: Generates random solution list of the specified length
    Parameters: start_value & end_value indicates the range of bins
    BPP1: Number of bins is 10
    BPP2: Number of bins is 100"""
    
    spawned_candidate_solutions_list = []
    for range_of_population in range(candidate_soln_population_size):
        result=[str(np.random.randint(start_value, end_value)) for digit in range(length_of_candidate)]
        joined_solution=" ".join(result)
        spawned_candidate_solutions_list.append(joined_solution)
    print("spawned candidate solution::",spawned_candidate_solutions_list)
    return spawned_candidate_solutions_list

def determine_candidate_solution_bins_weight(spawned_candidate_solution_list, mutation,problem_type):
    """Method_name:determine_candidate_solution_bins_weight()
    Functionality:
    Parameters:"""
    candidate_solution_bin_weight_dict = {}
    dict_i =0
    if problem_type == 'bpp1':
        #print("Problem statement::", problem_type)
        for solution in spawned_candidate_solution_list:
            i_n= 1
            bin_weight_dictionary = {}
            solution_list = list(solution.split(' '))
            for item in solution_list:
                if item in bin_weight_dictionary.keys():
                    existing_weight = bin_weight_dictionary[item]
                    bin_weight_dictionary[item] = existing_weight+i_n/2
                else:
                    bin_weight_dictionary[item] = i_n/2
                i_n =i_n+1
            dict_i+=1
            if mutation == True:
                candidate_solution_bin_weight_dict["candidate_solution_"+str(dict_i)+"m::"+solution] = bin_weight_dictionary
            else:
                candidate_solution_bin_weight_dict["candidate_solution_"+str(dict_i)+"::"+solution] = bin_weight_dictionary
    elif problem_type == 'bpp2':
        #print("Problem_type::", problem_type)
        for solution in spawned_candidate_solution_list:
            i_n= 1
            bin_weight_dictionary = {}
            solution_list = list(solution.split(' '))
            for item in solution_list:
                if item in bin_weight_dictionary.keys():
                    existing_weight = bin_weight_dictionary[item]
                    bin_weight_dictionary[item] = existing_weight+(i_n*i_n)/2
                else:
                    bin_weight_dictionary[item] = (i_n*i_n)/2
                i_n =i_n+1
            dict_i+=1
            if mutation == True:
                candidate_solution_bin_weight_dict["candidate_solution_"+str(dict_i)+"m::"+solution] = bin_weight_dictionary
            else:
                candidate_solution_bin_weight_dict["candidate_solution_"+str(dict_i)+"::"+solution] = bin_weight_dictionary        
    print("Candidate Solution bin Weight Dict::",candidate_solution_bin_weight_dict)
    return candidate_solution_bin_weight_dict
        
def evaluate_fitness_candidate_solution(candidate_solution_bin_weight_dict,problem_type):
    """Method_name:evaluate_fitness_candidate_solution()
    Functionality:
    Parameters:"""
    candidate_fitness_dictionary = {}
    if problem_type =='bpp1':
        b = 11
    elif problem_type == 'bpp2':
        b = 101
    else:
        pass
    for candidate_solution, bin_weight in candidate_solution_bin_weight_dict.items():
        print("Candidate Solution::", candidate_solution)
        bin_values = bin_weight.values()
        print("Bin Values::",bin_values)
        maximum_bin_value = max(bin_values)
        bin_list = list(bin_weight.keys())
        int_bin_list=[int(element) for element in bin_list]
        missing_bins = sorted(set(range(1,b))-set(int_bin_list))
        print("Missing bins::", missing_bins)
        if missing_bins:
            minimum_bin_value = 0.0
        else:
            minimum_bin_value = min(bin_values)
        fitness = maximum_bin_value - minimum_bin_value
        candidate_fitness_dictionary[candidate_solution] = fitness
    print("Candidate fitness dictionary::",candidate_fitness_dictionary)
    return candidate_fitness_dictionary

def choose_random_candidate_solution(candidate_solution_list,problem_type):
    """Method_name:evaluate_fitness_candidate_solution()
    Functionality:
    Parameters:
    Note: List is used bearing in mind the fact that list is ordered and dictionary as well(from 3.7)"""
    random_efficient_parent_dictionary = {}
    random_candidate_bin_weight_dict = {}
    parent_list = []
    i=1
    while i<3:
        #random.shuffle(candidate_solution_list)
        selected_random_candidate_solution_list = []
        random_candidate_solution1 = random.choice(candidate_solution_list)
        random_candidate_solution2 = random.choice(candidate_solution_list)
        selected_random_candidate_solution_list = selected_random_candidate_solution_list+[random_candidate_solution1,random_candidate_solution2]
        print("Selected Random Candidate Solution list::",selected_random_candidate_solution_list)
        selected_random_candidate_bin_weight_dict= determine_candidate_solution_bins_weight(selected_random_candidate_solution_list,False,problem_type)
        print("Selected Random Bin Weight::",selected_random_candidate_bin_weight_dict)
        fitness_candidate_solution_dictionary = evaluate_fitness_candidate_solution(selected_random_candidate_bin_weight_dict,problem_type)
        print("Fitness Candidate Soluton dict::",fitness_candidate_solution_dictionary)
        efficient_parent_solution = min(fitness_candidate_solution_dictionary, key = fitness_candidate_solution_dictionary.get)
        print("Efficient parent solution::",efficient_parent_solution)
        solution_to_parent_list = method_to_split(efficient_parent_solution)
        parent_list.append(solution_to_parent_list)
        i=i+1
    print("Parent list::", parent_list)
    return parent_list

def method_to_split(solution):
    """Method_name:method_to_split()
    Functionality:
    Parameters:"""
    string_1 = solution
    result = string_1.rpartition(string_1.rpartition(':')[1])
    return result[2]
        
def single_point_crossover(parent_list,length_of_candidate):
    """Method_name:single_point_crossover()
    Functionality:
    Parameters:"""
    random_cross_over_point = random.randint(1, length_of_candidate)
    print("Random Cross over point::", random_cross_over_point)
    split_parent_list = []
    for parent in parent_list:
        parent_solution_list = list(parent.split(' '))
        parent_first_half_substring = parent_solution_list[0:random_cross_over_point]
        parent_second_half_substring = parent_solution_list[random_cross_over_point:]
        split_parent_list = split_parent_list+[parent_first_half_substring,parent_second_half_substring]
    print("Split Parent List:",split_parent_list)
    cross_over_child_c = split_parent_list[0]+split_parent_list[3]
    cross_over_child_d = split_parent_list[2]+split_parent_list[1]
    print("Cross over child C and Cross over child D are {}, {} respectively.".format(cross_over_child_c,cross_over_child_d))
    return cross_over_child_c,cross_over_child_d

def mutation_at_a_random_point(cross_over_child_c,cross_over_child_d, k_operator,length_of_candidate):
    """Method_name:mutation_at_a_random_point()
    Functionality:
    Parameters:"""
    mutated_solution_list = []
    i = 1
    while i<=k_operator:
        print("Mutation size::", k_operator)
        random_mutation_point = random.randint(1, length_of_candidate)
        print("Random Mutation Point::", random_mutation_point)
        random_mutation_value = str(random.randint(1, 10))
        print("Random Mutation Value::",random_mutation_value)
        cross_over_child_c[random_mutation_point-1] = random_mutation_value
        #Mutation Again!
        random_mutation_point = random.randint(1, length_of_candidate)
        print("Random Mutation Point::", random_mutation_point)
        random_mutation_value = str(random.randint(1, 10))
        print("Random Mutation Value::",random_mutation_value)
        cross_over_child_d[random_mutation_point-1] = random_mutation_value
        mutated_child_c = ' '.join(cross_over_child_c)
        mutated_child_d = ' '.join(cross_over_child_d)
        mutated_solution_list = [] #Resetting list
        mutated_solution_list = mutated_solution_list+[mutated_child_c,mutated_child_d]
        i=i+1
    print("Mutated Solution List::",mutated_solution_list)
    return mutated_solution_list
    
def run_e_and_f_weakest_replacement(spawned_solution_list, mutated_fitness_dictionary, fitness_dictionary):
    """Method_name:run_e_and_f_weakest_replacement()
    Functionality:
    Parameters:"""
    """Logic: Iterate through fitness dictionary and get the object with high value score
    Iterate through mutated_fitness_dictionary and compare the elements with object with highest fitness score."""
    for mutated_element, mutated_element_value in mutated_fitness_dictionary.items():
        candidate_with_max_value = max(fitness_dictionary, key = fitness_dictionary.get)
        print("Candidate with maximum value::",candidate_with_max_value)
        if fitness_dictionary[candidate_with_max_value]>mutated_element_value:
            solution_to_remove = method_to_split(candidate_with_max_value)
            print("Solution to Remove::", solution_to_remove)
            popped_value = fitness_dictionary.pop(candidate_with_max_value)
            print("Popped Value::",popped_value)
            fitness_dictionary[mutated_element] = mutated_element_value
            solution_to_replace = method_to_split(mutated_element)
            spawned_solution_list.remove(solution_to_remove)
            spawned_solution_list.append(solution_to_replace)
            print("Solution list")
        else:
            print("Dont replace")
            pass
    return spawned_solution_list,fitness_dictionary

def plot_graph(loop_v,fitness_dictionary):
    """Plot graph between number of iterations and fitness indicating best fitness and average fitness"""
    """Method_name:run_e_and_f_weakest_replacement()
    Functionality:
    Parameters:"""
    average_fitness_list = []
    for candidate,fitness in fitness_dictionary.items():
        average_fitness_list.append(fitness)
    iterations = [i for i in range(1,loop_v)]
    average_maximum = max(average_fitness_list)
    fitness_value_arrived = min(average_fitness_list)
    print("Fitness value arrived::",fitness_value_arrived)
    plt.ylabel("Fitness")
    plt.xlabel("Number of Iterations")
    plt.xlim(0,loop_v)
    plt.ylim(0,average_maximum*0.75)
    plt.plot(iterations, average_fitness_list)
    plt.show()
    
def calculate_average_fitness(fitness_dictionary,average_fitness_dictionary,average_fitness_incrementor):
    """Method_name:calculate_average_fitness()
    Functionality:
    Parameters:"""
    afitness_list = []
    for candidate,fitness in fitness_dictionary.items():
        afitness_list.append(fitness)
    print("Fitness list::",afitness_list)
    average_fitness = sum(afitness_list)/len(afitness_list)
    average_fitness_dictionary["iteration::"+str(average_fitness_incrementor)] = average_fitness
    return average_fitness_dictionary

"""Initialization and Invocation of the methods starts here!"""
    
start_value=1
loop_v = 1
average_fitness_dictionary = {}
average_fitness_incrementor = 0
length_of_candidate = 10

candidate_soln_population_size = int(input("Please enter the candidate population size (10 or 100)::"))
mutation_size = int(input("Please enter the mutation value (1 or 5)::"))
while True:
    try:
        problem_type = input("Please enter the problem type to run i.e., bpp1 or bpp2")
        problem_type = problem_type.lower()
    except ValueError:
        print("Please check if the value is correct")
        continue

    if problem_type == "bpp1" or problem_type == "bpp2" :
        break
    else:
        print("Please enter appropriate value (either bpp1 or bpp2)")
if problem_type == 'bpp1':
    end_value = 11
else:
    end_value = 101
    
"""Parameters to change for trials"""    
cross_over_required = False
mutation_required = True
print("Methods are being invoked")

spawned_candidate_solution_list = spawn_candidate_solutions(start_value,end_value,length_of_candidate,candidate_soln_population_size)
print("Spawned Candidate Solution List::",spawned_candidate_solution_list)
candidate_solution_bin_weight_dict = determine_candidate_solution_bins_weight(spawned_candidate_solution_list,False,problem_type)
print("Candidate solution weight::",candidate_solution_bin_weight_dict)
fitness_dictionary = evaluate_fitness_candidate_solution(candidate_solution_bin_weight_dict,problem_type)
print("Fitness dictionary")
while loop_v<=1:
    parent_list = choose_random_candidate_solution(spawned_candidate_solution_list,problem_type)
    """Cross over to be made False for one of the trials"""
    if cross_over_required == True:
        print("Cross Over True")
        cross_over_child_c, cross_over_child_d = single_point_crossover(parent_list,length_of_candidate)
    else:
        print("Cross over False")
        cross_over_child_c = parent_list[0]
        cross_over_child_d = parent_list[1]
    """Mutation to be made False for one of the trials"""
    if mutation_required == True:
        print("Mutation True")
        mutated_solution_list = mutation_at_a_random_point(cross_over_child_c,cross_over_child_d,mutation_size,length_of_candidate)
    else:
        print("Mutation False")
        mutated_solution_list=[]
        mutated_solution_list = mutated_solution_list+[cross_over_child_c,cross_over_child_d]
    mutated_solution_bin_weight_dict = determine_candidate_solution_bins_weight(mutated_solution_list,True,problem_type)
    print("Mutated Solution bin Weight dictionary::",mutated_solution_bin_weight_dict)
    mutated_fitness_dictionary = evaluate_fitness_candidate_solution(mutated_solution_bin_weight_dict,problem_type)
    print("Mutated fitness dictionary::",mutated_fitness_dictionary)
    fitness_dictionary_reference = fitness_dictionary.copy()
    spawned_solution_list_reference = spawned_candidate_solution_list.copy()
    spawned_candidate_solution_list, fitness_dictionary = run_e_and_f_weakest_replacement(spawned_candidate_solution_list, mutated_fitness_dictionary, fitness_dictionary)
    print("Spawned Candidate solution list::", spawned_candidate_solution_list)
    print("Fitness dictionary::", fitness_dictionary)
    average_fitness_incrementor=average_fitness_incrementor+1   
    average_fitness_dictionary=calculate_average_fitness(fitness_dictionary,average_fitness_dictionary,average_fitness_incrementor)
    loop_v = loop_v+1

plot_graph(loop_v,average_fitness_dictionary)
```
